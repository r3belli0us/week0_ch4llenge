```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ethiopia = pd.read_csv("ethiopia_clean.csv")
kenya = pd.read_csv("kenya_clean.csv")
sudan = pd.read_csv("sudan_clean.csv")
tanzania = pd.read_csv("tanzania_clean.csv")
nigeria = pd.read_csv("nigeria_clean.csv")
```

```python
ethiopia["Country"] = "Ethiopia"
kenya["Country"] = "Kenya"
sudan["Country"] = "Sudan"
tanzania["Country"] = "Tanzania"
nigeria["Country"] = "Nigeria"
```

```python
df = pd.concat([ethiopia, kenya, sudan, tanzania, nigeria])
```

```python
df.groupby("Country")["T2M"].mean().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Country", y="T2M")
plt.title("Average Temperature by Country")
plt.show()
```

```python
df.groupby("Country")["PRECTOTCORR"].mean().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Country", y="PRECTOTCORR")
plt.title("Average Rainfall by Country")
plt.show()
```

```python
df["Extreme_Heat"] = df["T2M_MAX"] > 35

df.groupby("Country")["Extreme_Heat"].sum().sort_values()
```

```python
df.groupby("Country")["Extreme_Heat"].sum().plot(kind="bar")
plt.title("Extreme Heat Days by Country")
plt.show()
```

```python
df.groupby("Country")["RH2M"].mean().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Country", y="RH2M")
plt.title("Average Humidity by Country")
plt.show()
```

```python
df.groupby("Country")["WS2M"].mean().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Country", y="WS2M")
plt.title("Average Wind Speed by Country")
plt.show()
```

```python
df.groupby("Country")["T2M_RANGE"].mean().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Country", y="T2M_RANGE")
plt.title("Temperature Variability by Country")
plt.show()
```
