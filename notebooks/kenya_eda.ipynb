```python
import numpy as np

df.replace(-999, np.nan, inplace=True)
df.isna().sum()
```

Replaces -999 (which represents a missing data) with NaN so Python can properly treat it as a missing value.

```python
(df == -999).sum()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["T2M"])
plt.title("Temperature Over Time")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.show()
```

The graph shows that there isn't a massive long-term increase or decrease. So, we can conclude that the climate is pretty consistent in the 11 year period.

```python
plt.plot(df["PRECTOTCORR"])
plt.title("Rainfall Over Time")
plt.xlabel("Days")
plt.ylabel("Precipitation")
plt.show()
```

This shows several massive spikes, indicating the occurences of intense storm events. But on most days, the data suggests that there is little to no rain.
