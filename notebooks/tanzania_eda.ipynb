```python
import numpy as np

df.replace(-999, np.nan, inplace=True)
df.isna().sum()
```

Replaces -999 (which represents a missing data) with NaN so Python can properly treat it as a missing value.

```python
(df == -999).sum()
```

This is used to double-check for any fake missing values (-999).

```python
import matplotlib.pyplot as plt

plt.plot(df["T2M"])
plt.title("Temperature Over Time")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.show()
```

The temperature stays consistently warm. The climate remains stable without any extreme cold periods.

```python
plt.plot(df["PRECTOTCORR"])
plt.title("Rainfall Over Time")
plt.xlabel("Days")
plt.ylabel("Precipitation")
plt.show()
```

The graph suggests that this country experiences frequent and intense storm events.
